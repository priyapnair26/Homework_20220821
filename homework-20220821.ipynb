```python
# 10.If a list has repeating element or not
#x=[1,2,2,3,3,4,5,6,7,7]
x=[1,2,3,4,5,6,7]
x_length = len(x)
y=set(x)
y_length = len(y)

if x_length != y_length:
    print("List has repeating elements")
else:
    print("List does not have repeating elements")
```

```python
# 11.If two list has common elements or not
#x=[1,2,3,4,5,6,7]
#y=[1,2,3,8,9,10]
x=[1,2,3,4,5,6,7]
y=[8,9,10]
x=set(x)
y=set(y)
z=x.intersection(y)

if len(z) > 0:
    print("Both lists have common elements")
else:
    print("Both lists do not have common elements")
```
